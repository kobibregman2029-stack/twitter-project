```python
users = {}
posts = []
post_id = 1

def mainmenu():
    while True:
        print('Main Menu')
        print('#1: Create a user')
        print('#2: Make a post')
        print('#3: View your feed')
        print('#4: Like the posts')
        print('#5: Exit')
        answer = input('Enter a number 1-5: ')
        print('--------------------')
        if answer == '1':
            createuser()
        elif answer == '2':
            makeapost()
        elif answer == "3":
            viewfeed()
        elif answer == '4':
            likeapost()
        elif answer == '5':
            print('Goodbye')
            break
        else:
            print('Try Again bro')

def createuser():
    username = input("Pick Your Username: ")
    if username in users:
        print("That username is already taken")
        return
    bio = input("Write a short bio: ")
    users[username] = {"bio": bio, "followers": 0}
    print("Your user is created")

def makeapost():
    global post_id
    username = input("What is your username: ")
    if username in users:
        text = input("Type your post here: ")
        post = {"id": post_id, "user": username, "text": text, "likes": 0}
        posts.append(post)
        post_id = post_id + 1
        print("Your post is uploaded! ")
    else:
        print("That user does not exist, create a user first in to post")

def viewfeed():
    if len(posts) == 0:
        print("You have no posts up")
    for post in posts:
        print("Post #" + str(post["id"]))
        print(post["user"] + ": " + post["text"])
        print("Likes: " + str(post["likes"]))
        print()

def likeapost():
    number = input("Type the id of the post you would like to like: ")
    found = False
    for post in posts:
        if str(post["id"]) == number:
            post["likes"] = post["likes"] + 1
            print("You liked post #" + number)
            found = True
    if found == False:
        print("There is no post with that id")
mainmenu()
```
